## weighted_score = compound \* lg(favorite_count \* 2 + retweet_count) * lg(followers_count)
remove rows with compound=0

```python
import pandas as pd
```

```python
df = pd.read_csv('data/verizon.csv', dtype={'created_at':str, 'id_str':str})
y = pd.read_csv('data/verizon_price_change.csv', dtype={'time1':str, 'time2':str})
```

```python
df = df[df['compound'] != 0]
df = df.loc[(df['created_at'] >= '02281550') & (df['created_at'] < '03081550')]
df = df.reset_index(drop=True)
df
```

```python
from math import log10

def weighted(favorite, retweet, followers, compound):
    return compound * log10(favorite*2+retweet+1) * log10(followers+1)
```

```python
df['weighted_score'] = df.apply(lambda x: weighted(x['favorite_count'], x['retweet_count'], x['followers_count'], x['compound']), axis=1)
# df = df.loc[:, ['created_at','id_str','truncated','label','weighted_score']]
# df

df = df.loc[:, ['created_at','favorite_count','retweet_count','followers_count','compound','weighted_score']]
df
```

## Features

```python
# ex. 9:30     9:40     9:50    10:00
#    starts1  starts2  starts3  ends

tmp = ['0930','0940','0950','1000','1010','1020','1030','1040','1050','1100',
       '1110','1120','1130','1140','1150','1200','1210','1220','1230','1240',
       '1250','1300','1310','1320','1330','1340','1350','1400','1410','1420',
       '1430','1440','1450','1500','1510','1520','1530','1540','1550']
date = ['0301','0304','0305','0306','0307','0308']
        
ends = []
for d in date:
    for t in tmp[3:]:
        ends.append(d + t)

starts3 = []
for d in date:
    for t in tmp[2:-1]:
        starts3.append(d + t)
        
starts2 = []
for d in date:
    for t in tmp[1:-2]:
        starts2.append(d + t)

starts1 = []
for d in date:
    for t in tmp[:-3]:
        starts1.append(d + t)
```

```python
def features_price(start, end):
    prev = y.loc[y['time1'] == start]['price1'].tolist()[0]
    curr = y.loc[y['time2'] == end]['price2'].tolist()[0]
    delta = (curr - prev) * 10000 / prev
    return [round(delta, 1)]

from statistics import mean, median

def features_tweet(start, end):
    df_sub = df.loc[(df['created_at'] >= start) & (df['created_at'] < end)].sort_values(by=['weighted_score'])
    shift = len(df_sub) / 3
    if shift == 0: return ['drop']
    features = []
    features.extend(df_sub.loc[df_sub['compound'].idxmax()].tolist()[1:])
    features.extend(df_sub.loc[df_sub['compound'].idxmin()].tolist()[1:])
    features.extend([mean(df_sub['compound']), mean(df_sub['weighted_score']), mean(df_sub['favorite_count']), mean(df_sub['retweet_count']), mean(df_sub['followers_count']), median(df_sub['compound'])])
    features.extend([mean(df_sub[-shift:]['compound']), mean(df_sub[-shift:]['favorite_count']), mean(df_sub[-shift:]['retweet_count']), mean(df_sub[-shift:]['followers_count'])])
    features.extend([mean(df_sub[:shift]['compound']), mean(df_sub[:shift]['favorite_count']), mean(df_sub[:shift]['retweet_count']), mean(df_sub[:shift]['followers_count'])])
    shift_fa = min(shift, sum(df_sub['favorite_count'] > 0))
    features.append(mean(df_sub.sort_values(by=['favorite_count'])[-shift_fa:]['weighted_score']))
    shift_re = min(shift, sum(df_sub['retweet_count'] > 0))
    features.append(mean(df_sub.sort_values(by=['retweet_count'])[-shift_re:]['weighted_score']))
    shift_fo = min(shift, sum(df_sub['followers_count'] > 0))
    features.append(mean(df_sub.sort_values(by=['followers_count'])[-shift_fo:]['weighted_score']))
    return features
```

```python
rows = []
for i in range(len(ends)):
    row = [ends[i], y.loc[y['time2'] == ends[i]]['delta'].tolist()[0]]
    row.extend(features_price(starts2[i], starts3[i]))
    row.extend(features_price(starts1[i], starts3[i]))
    tmp = features_tweet(starts3[i], ends[i])
    if tmp[0] == 'drop':
        continue
    else:
        row.extend(tmp)
    row.extend(features_tweet(starts1[i], ends[i]))
    rows.append(row)

# c:compound, fa:favorite_count, re:retweet_count, fo:followers_count, w:weighted
# med:median, big: 1/3 bigger, small: 1/3 small
# ex. c_big_fa -> c_big_fa_avg : the average of favorite_count for 1/3 bigger compound rows
names = ['end', 'y', 'delta', 'delta_window']
names.extend(['c_max_fa', 'c_max_re', 'c_max_fo', 'c_max', 'c_w_max'])
names.extend(['c_min_fa', 'c_min_re', 'c_min_fo', 'c_min', 'c_w_min'])
names.extend(['c_avg', 'c_w_avg', 'fa_avg', 're_avg', 'fo_avg', 'c_med'])
names.extend(['c_big_c', 'c_big_fa', 'c_big_re', 'c_big_fo'])
names.extend(['c_small_c', 'c_small_fa', 'c_small_re', 'c_small_fo'])
names.extend(['fa_big_c', 're_big_c', 'fo_big_c'])
names.extend((name + '_window') for name in names[4:])
df_out = pd.DataFrame(rows, columns = names)
df_out
```

```python
df_out.to_csv('verizon_train.csv', index=False)
```
